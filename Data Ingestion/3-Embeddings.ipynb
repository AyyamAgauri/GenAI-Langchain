Embeddings are the vector representation of Data Chunks created in second step. They are necessary as LLM models get to know similarity between two sentences based on their embeddings. Embeddings are generally stored in clusters in vector databases.

There are many ways to create embeddings but using Langchain, it is easier to create them using HuggingFace, Ollama, OpenAI

```python
# Loading the Environment variables (HuggingFace Token, Langchain API)
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_community.embeddings import OllamaEmbeddings
embeddings = OllamaEmbeddings(model='gemma2')
```

```python
e1 = embeddings.embed_documents(
    ['This Embedding is created by Ollama opensource model Gemma2']
)
```
